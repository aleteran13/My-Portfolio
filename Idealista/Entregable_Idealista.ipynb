1. ¿Cuáles son las propiedades que tienen al menos 2 baños, un ascensor y una plaza de
aparcamiento? Selecciona el ASSETID, BATHNUMBER, HASLIFT , HASPARKINGSPACE y PRICE

```python
sSQL = """ 

SELECT ASSETID, BATHNUMBER,HASLIFT,HASPARKINGSPACE,PRICE
FROM madrid
WHERE BATHNUMBER >= 2 
     AND HASLIFT = 1
     AND HASPARKINGSPACE=1
ORDER BY BATHNUMBER , PRICE asc

"""
pd.read_sql(sSQL, con = conn)
```

2. ¿Cuál es el precio promedio de las propiedades con terraza?

```python
sSQL = """ 

SELECT ROUND(AVG(PRICE),2) AS AVG_PRICE
FROM madrid
WHERE HASTERRACE = 1 AND PRICE IS NOT NULL


"""
pd.read_sql(sSQL, con = conn)
```

```python
sSQL = """ 

SELECT ROOMNUMBER, COUNT(*) AS PROPERTIES,
        ROUND(AVG(PRICE),2) AS AVG_PRICE_ROOMNUMBER
FROM madrid
GROUP BY ROOMNUMBER
ORDER BY ROOMNUMBER

"""
pd.read_sql(sSQL, con = conn)
```

10. ¿Hay alguna tendencia en el valor medio de las propiedades en los distintos meses del
año 2018? ¿Y en el valor máximo o mínimo?


```python
sSQL = """ 
SELECT  
    MIN(PRICE) AS MIN_PRICE,
    ROUND(AVG(PRICE), 2) AS AVG_PRICE,
    MAX(PRICE) AS MAX_PRICE,
    EXTRACT(MONTH FROM STR_TO_DATE(CONCAT(PERIOD, '01'), '%Y%m%d')) AS MONTH
FROM madrid
GROUP BY MONTH
ORDER BY MONTH

"""
pd.read_sql(sSQL, con = conn)
# el precio minimo se reduce en el primer semestre pero se estabiliza en el siguiente
#trimestre 
#el precio maximo se mantiene constante 
#el precio promedio decrece al principio del año pero al final crece de nuevo 
```

14. ¿Cuál es el precio promedio por metro cuadrado para propiedades con orientación sur,
este y oeste?


```python
sSQL = """ 


SELECT 
  ORIENTATION, ROUND(AVG(UNITPRICE), 2) AS AVG_PRICE_M2
FROM (
  SELECT 
    CASE 
      WHEN HASSOUTHORIENTATION = 1 THEN 'SOUTH'
      WHEN HASEASTORIENTATION = 1 THEN 'EAST'
      WHEN HASWESTORIENTATION = 1 THEN 'WEST'
    END AS ORIENTATION,
    UNITPRICE
  FROM madrid
 WHERE 
    (HASSOUTHORIENTATION = 1 OR HASEASTORIENTATION = 1 OR HASWESTORIENTATION = 1)
    AND UNITPRICE IS NOT NULL
) AS FILTRO
 
GROUP BY ORIENTATION
   
"""
pd.read_sql(sSQL, con = conn)
```

15. ¿Cuál es la diferencia de los precios promedios entre propiedades que incluyen la plaza
de
aparcamiento y las que no?

```python
sSQL = """ 

SELECT 
  ROUND(A.AVG_PRICE - B.AVG_PRICE, 2) AS DIFERENCIA
FROM (
        SELECT ROUND(AVG(PRICE),2) AS AVG_PRICE
        FROM madrid
        WHERE ISPARKINGSPACEINCLUDEDINPRICE = 1) A,
(
  SELECT AVG(PRICE) AS AVG_PRICE
  FROM madrid
  WHERE ISPARKINGSPACEINCLUDEDINPRICE = 0
) B

"""
pd.read_sql(sSQL, con = conn)
```

```python
sSQL = """ 
SELECT 
  ROUND(AVG(CASE WHEN ISPARKINGSPACEINCLUDEDINPRICE = 1 THEN PRICE END), 2) AS AVG_CON_PARKING,
  ROUND(AVG(CASE WHEN ISPARKINGSPACEINCLUDEDINPRICE = 0 THEN PRICE END), 2) AS AVG_SIN_PARKING,
  ROUND(
    AVG(CASE WHEN ISPARKINGSPACEINCLUDEDINPRICE = 1 THEN PRICE END) -
    AVG(CASE WHEN ISPARKINGSPACEINCLUDEDINPRICE = 0 THEN PRICE END), 2
  ) AS DIFERENCIA
FROM madrid
"""
pd.read_sql(sSQL, con = conn)
```

2. ¿Cuáles son todas las propiedades gestionadas por una agencia con puntuación mayor a
4.5? Muestra el ASSETID, AGENCY_NAME, PRICE y RATING.


```python

sSQL = """ 

SELECT m.ASSETID, a.AGENCY_NAME, m.PRICE, a.RATING
FROM madrid m
INNER JOIN AGENCIAS a
    ON m.AGENCY_ID = a.AGENCY_ID
WHERE a.RATING > 4.5
ORDER BY a.RATING DESC, m.PRICE DESC

    
"""
pd.read_sql(sSQL, con = conn)
```

4. ¿Cuál es la inmobiliaria con el precio promedio más alto en propiedades listadas?


```python
sSQL= """ 
SELECT a.AGENCY_NAME , ROUND(AVG(m.PRICE),2) AS PRECIO_AVG
FROM AGENCIAS a 
JOIN madrid m 
    ON a.AGENCY_ID = m.AGENCY_ID
GROUP BY a.AGENCY_NAME 
ORDER BY PRECIO_AVG DESC
"""
pd.read_sql(sSQL, con = conn)
```
